```python
import ast
def get_dataframe(files):
    columns = "actor_type,actor_id,action_type,action_time,object_type,object_id,channel_type,channel_id,attrs,partition,theme"
    columns = columns.split(",")

    cti_frames = []
    for filepath in glob.glob(files):
        results = []
        with open(filepath, "rb") as in_file:
            for i, line in enumerate(in_file):
                fields = line.strip().split(",")
                values = fields[:8] + [",".join(fields[8:-2])] + fields[-2:]

                results.append(dict(zip(columns, values)))

        df = pd.DataFrame(results)
        cti_frames.append(df)

    cti_df = pd.concat(cti_frames,ignore_index=True)
    cti_df["attrs"] = cti_df["attrs"].map(lambda x: ast.literal_eval(json.loads(x)))
    
    return cti_df
```

```python
cti_df = get_dataframe(os.path.join(cti_folder, "*csv"))
print cti_folder, cti_df.shape
    
cctxn_df = get_dataframe(os.path.join(cctxn_folder, "*csv"))
print cctxn_folder, cctxn_df.shape
    
atm_df = get_dataframe(os.path.join(atm_folder, "*csv"))
print atm_folder, atm_df.shape
    
mybank_df = get_dataframe(os.path.join(mybank_folder, "*csv"))
print mybank_folder, mybank_df.shape
```

### net_profit change & aum_net_profit_diff change through time

```python
def change_observe(observe_column):
    old = pd.DataFrame(data = {observe_column:old_df[observe_column],"time":["Old" for _ in range(len(old_df[observe_column]))]})
    new = pd.DataFrame(data = {observe_column:new_df[observe_column],"time":["New" for _ in range(len(new_df[observe_column]))]})
    frames = [old,new]
    merge = pd.concat(frames, ignore_index=True)
    sns.boxplot(data = merge, x = "time", y = observe_column)
    plt.show()
```
